#### Importing Data

Import the dataset, assign column names and view the data

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, header = None)
headers =["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors", "body_style", "drive-wheels", "engine-location", "wheel_base", "length", "width", "height", "curb-weight", "engine_type", "num-of-cylinders", "engine-size", "fuel-system", "bore", "stroke", "compression_ration", "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"]
df.columns = headers
df.head(10)
```

We see **?** in the data, let's replace it with **NaN** for a cleaner look.

```python
df1 = df.replace('?', np.nan)
df1.head(10)
```

Since we're interested in used car prices, we'll want to ensure the **price** data type is correct.

#### Data Normalization

Because one day we may want to model this data for machine learning, we should normalize some of the columns so they don't have more weight than other similar columns.

There are a few ways to do this, so it's important to assess the data you have. Generally, I use **simple feature scaling** which is...  

(orginal_value / max of that column).

```python
df1["length"] = df1["length"] / df1["length"].max()

df1["width"] = df1["width"] / df1["width"].max()

df1["height"] = df1["height"] / df1["height"].max()

df1[["length","width","height"]].head()
```

#### Data Manipulation

Since I'm in Canada, I'd like to see the **city-mpg** column as **city-litres/100km**.

```python
df1['city-litres/100km'] = 235 / df1['city-mpg'] # convert the values and create another column
df1.head()
```

Identify and handle missing values

<u>Options</u>
1. Replace the data
2. Drop the record/column
3. Leave the missing data

```python
missing_data = df1.isnull()
missing_data.head(10)

#true = missing data
#false = data is not missing
```

Counting the missing values in each column to see how big of an issue it is.

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

The **normalized-losses** has a lot of missing values, and we don't need that column for this analysis, so let's drop the entire column.

```python
df1.drop("normalized-losses", axis=1, inplace=True)
df1.head()
```

#### Data Visualization

The main point of this analysis was to see which car attributes affected used car prices the most, we can use a **scatterplot** too see the correlations between variables. Let's look at engine size and how price it affected by it.

```python
y=df1['price']
x=df['engine-size']
plt.scatter(x,y)
plt.title('Engine Size vs. Price')
plt.xlabel('Engine Size')
plt.ylabel('Price')
```

Another useful visualization is a **scatterplot**, which we can use to see the distribution of prices for gas and diesel cars.

```python
sns.boxplot(x='fuel-type',y='price',data=df1)
```

Based on the two visualizations, there is a strong correlation between engine-size and used car price. Additionally, diesel cars have a higher median than gas cars, so you can typically expect to pay more for diesel cars over gas.
